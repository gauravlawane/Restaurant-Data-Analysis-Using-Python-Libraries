```python
# import all Libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset:
df = pd.read_csv(r"C:\Users\asus\OneDrive\Desktop\data analytic\intern Dataset.csv")
df.head(21)
```

```python
# numbers of rows and columns:
df.shape
```

```python
#check the column name and datatype:
df.info()
```

 1) Identify the city with the highest number of restaurants in the dataset.

```python
print(df['City'].value_counts()[[0]].index[0])
```

 2) Calculate the average rating for restaurants in each city.

```python
average_restaurant_ratings = df.groupby('City').agg(Average_Rating=('Aggregate rating', 'mean')).sort_values('Average_Rating', ascending=False)
print("Average ratings for restaurants in each city, sorted by average rating:")
print(average_restaurant_ratings)
```

```python
ax = average_restaurant_ratings.plot(kind='bar', color='#027DFF', edgecolor='black', figsize=(26, 8))
plt.xlabel('City')
plt.ylabel('Average Rating')
plt.title('Average Ratings for Restaurants in Each City')
plt.show()
```

3) Determine the city with the highest average rating.

```python
city_with_highest_rating = average_restaurant_ratings['Average_Rating'].idxmax()
highest_rating = average_restaurant_ratings['Average_Rating'].max()

print(f"The city with the highest average rating is {city_with_highest_rating} with an average rating of {highest_rating:.2f}")
```
